# This Tutorial is meant to get aquinted with the the soc.py class 

Although the Model of self-organized criticality can relatively simply be employed manually using brian2 (see the tutorial "Introduction" for a step-by step-guide), we also implemented a method to readily employ models by specifying the desired parameters.

```python
import soc
```

The most important parameters of the model are the network size N, the critical parameter alpha and the duration of the simulation. For all other parameters one can select, please refer to the soc.py class.

```python
random_state = 42
N = 300
duration = 100000
alpha = 1.4
```

We can simply pass these parameters to the build_and_run function. By default a raster plot will be plotted after the simulation.

```python
states, spikes = soc.build_and_run(duration, N, alpha, random_state = random_state)
```

We can pass the spike data, collected in the brian2.SpikeMonitor class to the avalanche distribution method. Which creates a log-log plot of the avalanche distribution.

```python
soc.avalanche_distribution(spikes)
```
